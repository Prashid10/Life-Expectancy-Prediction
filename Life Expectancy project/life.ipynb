```python
# IQR (Interquartile range)
# Except schooling other features are not in normal distribution so we use IQR method to trim or capp.
def cap_outliers(data,feat):
    q1=data[feat].quantile(0.25)
    q3=data[feat].quantile(0.75)
    IQR=q3-q1
    ub=q3+1.5*IQR
    lb=q1-1.5*IQR
    data[feat]=np.clip(data[feat],min=lb,max=ub)
    return data
```

```python
for i in numerical_columns:
    cap_outliers(data,i)
```

## RÂ² Values Interpretation:

- RÂ² > 0.8:  
  Good model (explains over 80% of the variance). The model fits the data quite well.

- RÂ² between 0.6 and 0.8:  
  Decent model, but there may still be room for improvement.

- RÂ² between 0.4 and 0.6:  
  Mediocre model. The model does not explain much of the variance, and improvements might be needed.

- RÂ² < 0.4:  
  Poor model. Significant improvements should be considered, possibly by choosing a better model or improving the data preprocessing.

- Negative RÂ²:  
  The model is worse than predicting the mean of the target variable. Major changes are likely needed, such as switching to a different model, adjusting hyperparameters, or revisiting data quality.

```python
from scipy import stats
```

```python
num = ['Life expectancy ', 'Adult Mortality', 'Alcohol', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population', ' thinness 5-9 years',
       'Schooling']
```

```python
def pearson_test(data, num, target = 'Life expectancy '):
    for i in num: 
        x_num = data[i]
        y_num = data[target]
   
        r, p = stats.pearsonr(x_num, y_num)
        print(f'Pearson correlation between {i} and {target}: r = {r:.4f}, p-value = {p:.4f}')
```

```python
pearson_test(data, num)
```
